- Mainly used for `making decisions`
- `if`, `elif`, `else` - used to chooses between options
- Python relies on **indentation** (whitespace at the beginning of a line) to **define scope** in the code. Other programming languages often use curly-brackets for this purpose.

#### IF statement
- An "if statement" is written by using the `if` keyword.
- Syntax
```
    ... main_code

    if (condition):
        if_code1
        if_code2
        ...

    main_code ...
```
- **How it works?** - The if statement evaluates a condition (an expression that results in `True` or `False`). If the **condition is `True`**, the code block inside the **if statement is executed**. If the **condition is False**, the **code block is skipped**.

````python
# Example 1
age = 18
print("Question: Is the person eligible to vote?")
print("*"*80)
if age>=18:
    print(f"Since the age is {age}, the person is eligible to vote")
print("*"*80)
````

#### Elif Statement
- The `elif` keyword is Python's way of saying "if the previous conditions were not true, then try this condition".
- Use elif when you have multiple mutually exclusive conditions to check. This is more efficient than using multiple separate if statements because Python stops checking once it finds a true condition.
- Syntax
```
    ... main_code

    if (condition):
        if_code1
        if_code2
        ...
    elif (condition):
        elif_code1
        elif_code2
        ...

    main_code ...
```

````python
score = 75

if score >= 90:
  print("Grade: A")
elif score >= 80:
  print("Grade: B")
elif score >= 70:
  print("Grade: C")
elif score >= 60:
  print("Grade: D")
````

#### Else Statement
- The `else` keyword catches anything which isn't caught by the preceding conditions.
- The else` statement is executed when the `if` condition (and any `elif` conditions) evaluate to **False**.

````python
# Example 1
age = 16
print("Question: Is the person eligible to vote?")
print("*"*80)
if age>=18:
    print(f"Since the age is {age}, the person is eligible to vote")
else:
    print(f"Since the age is {age}, the person is not eligible to vote")
print("*"*80)
````

````python
score = 50

if score >= 90:
  print("Grade: A")
elif score >= 80:
  print("Grade: B")
elif score >= 70:
  print("Grade: C")
elif score >= 60:
  print("Grade: D")
else:
    print("Grade F")
````

#### Shorthand conditional statements
- If you have only one statement to execute, you can put it on the same line as the if statement.<br>
***Note:*** You still need the colon : after the condition.

- If you have one statement for if and one for else, you can put them on the same line using a **conditional expression**

- Shorthand if statements and ternary operators should be used when:
    - The condition and actions are simple
    - It improves code readability
    - You want to make a quick assignment based on a condition<br>


***Important***: While shorthand if statements can make code more concise, avoid overusing them for complex conditions. For readability, use regular if-else statements when dealing with multiple lines of code or complex logic.

````python
a = 5
b = 2
if a > b: print("a is greater than b")
````

````python
a = 2
b = 330
print("A") if a > b else print("B")
````

````python
#Assign a Value With If ... Else
a = 10
b = 20
bigger = a if a > b else b
print("Bigger is", bigger)
````

````python
# Multiple Conditions on One Line
a = 330
b = 330
print("A") if a > b else print("Equal") if a == b else print("B")
````

#### Logical Operators
Logical operators are used to combine conditional statements. Python has three logical operators:
- `and` - Returns **True** if **both** statements are True
- `or` - Returns **True** if **one of the statements** is True
- `not` - **Reverses** the result, returns False if the result is True

````python
# AND
a = 200
b = 33
c = 500
if a > b and c > a:
  print("Both conditions are True")
````

````python
# OR
a = 200
b = 33
c = 500
if a > b or a > c:
  print("At least one of the conditions is True")
````

````python
# NOT
a = 33
b = 200
if not a > b:
  print("a is NOT greater than b")
````

````python
age = 25
is_student = False
has_discount_code = True

if (age < 18 or age > 65) and not is_student or has_discount_code:
  print("Discount applies!")
````

#### Nested Conditional Statements
You can have `if` statements inside `if` statements. This is called nested `if` statements.

````python
age = 25
has_license = True

if age >= 18:
  if has_license:
    print("You can drive")
  else:
    print("You need a license")
else:
  print("You are too young to drive")
````

````python
score = 85
attendance = 90
submitted = True

if score >= 60:
  if attendance >= 80:
    if submitted:
      print("Pass with good standing")
    else:
      print("Pass but missing assignment")
  else:
    print("Pass but low attendance")
else:
  print("Fail")
````

#### Pass Statement
`if` statements **cannot be empty**, but if you for some reason have an `if` statement with **no content**, put in the **pass** statement to **avoid** getting an **error**.

#### Why Use pass?
The pass statement is useful in several situations:
- When you're creating code structure but haven't implemented the logic yet
- When a statement is required syntactically but no action is needed
- As a placeholder for future code during development
- In empty functions or classes that you plan to implement later


````python
a = 33
b = 200

if b > a:
  pass
````
