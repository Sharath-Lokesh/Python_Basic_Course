### Variables

Variables are containers for storing data values.

Think of variables as *cookie jar* and values as *cookies*.
One can store *chocolates or pickel or grains* etc. inside the *cookie jar*, likewise **variables** can also store **different types of values** (integers, decimal, letters, strings etc.)

A variable is created the moment you first assign a value to it.

**Syntax**
    `variable = value`

```python
x = 5 # x is a variable containing an interger value 5
y = "=John" # y is a variable containing a string value John
print(x) # print the value stored in x
print(y) # print the value stored in y
```

Variables do not need to be declared with any particular type, and can even change type after they have been set.

```python
x = 4       # x is of type integer
print(x)
x = "Sally" # x is now of type string
print(x)
```

```python
x=4
x="sally" # over-writing a variable
print(x)
```

### Casting

If you want to specify the data type of a variable, this can be done with casting.
Converting one data type to another.

Casting is like changing the type of cookie (chocochip, almond etc.) you’re putting into the jar. Casting is like repacking or converting what’s inside into a different kind so it can be used the right way.


#### Get the Type
You can get the data type of a variable with the `type()` function.

```python
x = str(3)    # x will be '3' x="3"
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0
print("x: ", x, "type of x: ", type(x))
print("y: ", y, "type of y: ", type(y))
print("z", z, "type of z: ", type(z))
```

#### Single or Double Quotes?
String variables can be declared either by using single or double quotes:

```python
x = "John"
print(type(x))
# is the same as
x = 'John'
print(type(x))
```

#### Case-Sensitive
Variable names are case-sensitive.

```python
a = 4
A = "Sally"
#A will not overwrite a
print("a: ", a)
print("A: ", A)
```

### Variable Names

A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume).

Rules for Python variables:

- A variable name must start with a **letter** or the **underscore character**
- A variable name cannot start with a **number**
- A variable name can only contain **alpha-numeric** characters and underscores (A to z, 0 to 9, and _ )
- Variable names are **case-sensitive** (age, Age and AGE are three different variables)
- A variable name cannot be any of the **Python keywords**.

***NOTE:*** Remember that variable names are case-sensitive

```python
# int = 5 ## can't do this
y = int(6.0)
```

#### Multi Words Variable Names

Variable names with more than one word can be difficult to read.

There are several techniques you can use to make them more readable:

#### Camel Case
Each word, except the first, starts with a capital letter:
`personName = "John Snow"`

#### Pascal Case
Each word starts with a capital letter:
`PersonName = "John Snow"`

#### Snake Case
Each word is separated by an underscore character (`_`):
`person_name = "John Snow"`

### Assign Multiple Values

#### Many Values to Multiple Variables
Python allows you to assign values to multiple variables in one line:

```python
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)
```

#### One Value to Multiple Variables
You can assign the same value to multiple variables in one line:

```python
x = y = z = "Orange"
print(x)
print(y)
print(z)
```

### Scope of a variable

#### Global Scope
Variables that are created outside of a function are known as global variables.

Global variables can be used by everyone, both inside of functions and outside.

Think of gloabl variables as a *cookie jar with some cookies, kept in the living room*. *You* can have a cookie and your *family members* can also have a cookie.

#### Local Scope
Variables that are created inside of a function are known as local variables. It is used with in the function.

Think of local variables as a *cookie jar with some cookies, kept in your room*. Only you can have a cookie.

#### Global Keyword
Local variables can be made global with the use of `global` keyword. Cookie jar is in your romm but you inform others that they can have a cookie.

Also, use the global keyword if you want to change a global variable inside a function.

```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

```python
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

```python
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

```python
x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```
