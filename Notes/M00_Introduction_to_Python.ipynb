### Installing Anaconda 3 IDE - Launching Jupyter Notebook

1. Download Anaconda (Individual Edition) for your OS (Windows/macOS/Linux) from the official Anaconda website. https://www.anaconda.com/download

### Launching Jupyter Notebook

#### Opening a terminal in a desired folder, then start Jupyter
1. Go to the folder you want notebooks to live in.
2. Copy the path of the folder.
3. Open Anaconda Command Prompt.
4. Change directory to the desired folder and execute the following command `cd <paste the copied folder path>`.
5. Execute the command `jupyter notebook` in anaconda command prompt. 

### Why Jupyter Notebook

Jupyter Notebook is useful because it makes Python feel like an interactive lab notebook—especially good for learning and for science-style work.
1. **Immediate feedback**: run code in small blocks (“cells”) and see results right away without running an entire file.
2. **Great for teaching and learning**: you can explain step-by-step, then run the next step. Easy to pause, experiment, and correct mistakes.
3. **Mix code + notes + math**: combine Python code with plain-English explanations, images, and equations (helpful for physics/astronomy).
4. **Data and plotting workflow**: display tables, arrays, and plots inline (matplotlib), which is how scientific work is often explored.
5. **Encourages experimentation**: change one line, rerun one cell, compare outputs quickly.
6. **Easy to share**: a single .ipynb file can be shared with a student; it keeps the code and results together.

#### When not to use it:
- Building larger programs/apps with multiple files → use a normal .py project in an IDE.
- When you need clean “run from start to finish” behavior every time → scripts are usually better.

### Syntax

Syntax is the set of rules for how you write Python—*like grammar rules in English.*

Imagine Python is a robot that only understands a specific language.
If you say the sentence in the wrong way, the robot gets confused and stops.

**Why syntax is needed**
- It tells Python exactly what you mean
- It prevents confusion (computers don’t “guess”)
- It keeps code readable for humans too

### Comments
***Comments*** are notes you write in Python code for humans to read. Python ignores them when running the program.

#### Why comments are useful
- Explain why the code exists (the goal or reasoning)
- Make code easier to understand later (for you or others)
- Mark TODOs or tricky parts
- Help teaching and learning by describing steps

### Print and Input statements
Think of a Python program like a small robot.

#### print()
***print() = the robot talking***

- If the robot never talks, you won’t know what it did.
- It can tell you answers: “The result is 42”
- It can tell you what it’s doing: “Calculating…”
- It helps you check if it’s working: “I got this value: 10”

**What it does**: shows information to the user on the screen (console).

**Why it’s needed**:
- To see results of calculations
- To debug (check what values your variables have)
- To communicate what the program is doing

```python
print("Hello World!") # This is a print statement
```

#### input()
***input() = you talking to the robot***
- If the robot can’t listen, it can only do the same thing every time.
- You can give it your name
- You can tell it a number
- You can choose what it should do next

**What it does**: lets the program ask the user a question and receive what they type.

**Why it’s needed**:
- To make programs interactive
- To let the user provide values instead of hard-coding them
- To reuse the same program for different cases

```python
input("How old are you?")
```

```python
print(("My name is"), input("What is your name?"), input("age?")) # print("My name is", "sharath")
```
