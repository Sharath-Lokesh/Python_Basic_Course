## Python Collections (Arrays)
There are four collection data types in the Python programming language:
- `List` is a collection which is **ordered and changeable**. **Allows duplicate** members.
- `Tuple` is a collection which is **ordered and unchangeable**. **Allows duplicate** members.
- `Set` is a collection which is **unordered, unchangeable, and unindexed**. **No duplicate** members.
- `Dictionary` is a collection which is **ordered and changeable**. **No duplicate** members.

## List

**Lists are used to store multiple items in a single variable.**

- Lists are one of 4 built-in data types in Python used to store collections of data.
- Lists are created using square brackets: `[]`
- The values inside the list are called **List Items**. List items are ordered, changeable, and allow duplicate values.
- List items are indexed, the first item has index `[0]`
- If you add new items to a list, the new items will be placed at the end of the list. We can change, add, and remove items in a list after it has been created.

To determine how many items a list has, use the `len()` function:

```python
thislist = ["apple", "banana", "cherry"]
print(len(thislist))
```

A list can contain different data types:

```python
list1 = ["abc", 34, True, 40, "male"]
```

### The list() Constructor
It is also possible to use the list() constructor when creating a new list.

```python
list2 = list(("apple", "banana", "cherry")) # note the double round-brackets
print(list2)
```

#### Accessing items in a List
- List items are indexed and you can access them by referring to the index number. ***Note***: The first item has index `0`.
- Negative indexing means start from the end. `-1` refers to the last item, `-2`refers to the second last item etc.
- You can specify a **range of indexes** by specifying where to `start` and where to `end` (not included) the range. When specifying a range, the **return value** will be a **new list** with the specified items.
- By **leaving out the start value**, the range will start at the first item.
- By **leaving out the end value**, the range will go on to the end of the list.
- Specify negative indexes if you want to start the search from the end of the list.
- Syntax for list indexing `list_name[start_idx : end_idx : steps]` ***Note***: By default `steps=1`

```python
list1 = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
```

```python
print(list1[1]) # indexing
```

```python
print(list1[-1]) # negative indexing
```

```python
print(list1[2:5]) #range - include : start index :: exclude : end index
```

```python
print(list1[:5]) # no start index
```

```python
print(list1[2:]) # no end index
```

```python
print(list1[:])
print(list1)
```

```python
print(list1[-5:-3]) # negative index range
```

```python
print(list1[-5:]) # negative index range
```

```python
print(list1[:-3]) # negative index range
```

```python
print(list1[1::2]) # step = 2
```

## Tuple 

**Tuples are used to store multiple items in a single variable.**
- Tuple is one of 4 built-in data types in Python used to store collections of data.
- A tuple is a collection which is ordered and **unchangeable** (cannot change, add or remove items after the tuple has been created).
- Tuples are written with round brackets `()`.
- Tuple items are ordered, unchangeable, and allow duplicate values.
- Tuple items are indexed, the first item has index `[0]`

```python
tuple1 = ("apple", "banana", "cherry", "apple", "cherry")
print(tuple1)
print(len(tuple1))
```

```python
### Create Tuple With One Item
thistuple = ("apple",)
print(type(thistuple))

#NOT a tuple --> WHY??
thistuple = ("apple")
print(type(thistuple))
```

```python
tuple1 = ("abc", 34, True, 40, "male")
print(tuple1)
```

#### The tuple() Constructor
It is also possible to use the `tuple()` constructor to make a tuple.

```python
thistuple = tuple(("apple", "banana", "cherry")) # note the double round-brackets
print(thistuple)
```

#### Change Tuple Values
Once a tuple is created, you cannot change its values. Tuples are **unchangeable, or immutable** as it also is called.<br>
But there is a workaround. You can convert the **tuple into a list**, change the list, and convert the **list back into a tuple**.

```python
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi" # modify
y.append("orange") # add
y.remove("apple") #remove
x = tuple(y)

print(x)
```

#### Unpacking a Tuple
When we create a tuple, we normally assign values to it. This is called "packing" a tuple: <br>
***Note***: The number of variables must match the number of values in the tuple, if not, you must use an asterisk to collect the remaining values as a list.

```python
fruits = ("apple", "banana", "cherry") # packing

(green, yellow, red) = fruits #unpacking

print(green)
print(yellow)
print(red)
```

If the number of variables is less than the number of values, you can add an * to the variable name and the values will be assigned to the variable as a list:

```python
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = fruits

print(green)
print(yellow)
print(red) # creates a lsit
```

```python
fruits = ("apple", "mango", "papaya", "pineapple", "cherry")

(green, *tropic, red) = fruits

print(green)
print(tropic)
print(red)
```

```python
# join tuples
tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2
print(tuple3)
```

```python
# multiply tuples
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2

print(mytuple)
```

## Sets

- Sets are used to store multiple items in a single variable.
- Set is one of 4 built-in data types in Python used to store collections of data.
- A set is a collection which is **unordered**, **unchangeable** (only you can remove items and add new items), and **unindexed**.
- do not allow duplicate values.
- You cannot access items in a set by referring to an index or a key.

```python
set1 = {"apple", "banana", "cherry"}
print(set1) # Sets are unordered, so you cannot be sure in which order the items will appear.
```

```python
set1 = {"apple", "banana", "cherry", "apple"}
print(set1) # No duplicates
```

```python
# The values True and 1; False and 0 are considered the same value in sets, and are treated as duplicates:
set1 = {"apple", "banana", "cherry", True, 1, 2, False, 0}

print(set1)
```

```python
print(len(set1))
```

#### The set() Constructor
It is also possible to use the `set()` constructor to make a set.

```python
set1 = set(("apple", "banana", "cherry")) # note the double round-brackets
print(set1)
```

```python
set1.add("orange") # add item
print(set1)
```

```python
tropical = {"pineapple", "mango", "papaya"}
```

```python
set1.update(tropical) # add sets
print(set1)
```

```python
mylist = ["kiwi", "orange"]
set1.update(mylist) # update can take any iterable object
print(set1)
```

```python
mydict = {1:1, 2:4, 3:9}
set1.update(mydict) # update can take any iterable object
print(set1)
```

##### Remove Set Items
- If the item to remove does not exist, `remove()` will raise an error.
- If the item to remove does not exist, `discard()` will NOT raise an error.
- You can also use the `pop()` method to remove an item, but this method will remove a random item, so you cannot be sure what item that gets removed.
- The `clear()` method empties the set.
- The `del` keyword will delete the set completely.

```python
set1.remove("banana")

print(set1)
```

```python
set1.discard("banana")
print(set1)
```

```python
x = set1.pop()
print(x)
print(set1)
```

```python
set1.clear()
print(set1)
```

```python
set1 = {"apple", "banana", "cherry"}
print(set1)
del set1
print(set1)
```

#### Join Sets
There are several ways to join two or more sets in Python.
- The `union()` and `update()` methods joins all items from both sets.
- The `intersection()` method keeps ONLY the duplicates.
- The `difference()` method keeps the items from the first set that are not in the other set(s).
- The `symmetric_difference()` method keeps all items EXCEPT the duplicates.



```python
# Union
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)
print(set3)
```

```python
#union
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1 | set2
print(set3)
```

```python
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}

myset = set1.union(set2, set3, set4)
print(myset)
```

```python
# intersection
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.intersection(set2)
print(set3)
```

```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 & set2
print(set3)
```

```python
# difference
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.difference(set2)
set4 = set2.difference(set1)

print(set3)
print(set4)
```

```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 - set2
print(set3)
```

```python
# symmetric difference
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.symmetric_difference(set2)

print(set3)
```

```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 ^ set2
print(set3)
```

#### Frozenset
- frozenset is an immutable version of a set.
- Like sets, it contains unique, unordered, unchangeable elements.
- Unlike sets, elements cannot be added or removed from a frozenset.

```python
x = frozenset({"apple", "banana", "cherry"})
print(x)
print(type(x))
```
