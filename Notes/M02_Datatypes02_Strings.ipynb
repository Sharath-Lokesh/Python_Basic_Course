### Strings
- Strings in python are surrounded by either single quotation marks `''`, or double quotation marks `""`.
- `'hello'` is the same as `"hello"`
- can use **quotes inside a string**, as long as they don't match the quotes surrounding the string.

```python
print("Hello")
print('Hello')
```

```python
print("It's alright")
print("He is called 'Johnny'")
print('He is called "Johnny"')
```

```python
# Assigning a string to a variable is done with the variable name followed by an equal sign and the string:
a = "Hello"
print(a)
```

```python
# You can assign a multiline string to a variable by using three quotes:

# You can use three double quotes:

a = """Hey there!
I'm hello world A,
How are you?
"""
print(a)

# Or three single quotes:
b = '''Hey there!
I'm hello world B,
How are you?
'''
print(b)
```

### Strings are Arrays

#### What is an array?
An array is like a row of boxes in a line.
- Each box holds one item
- Every box has a number (position) starting from 0
- You can go to any box directly using its number

#### What does “Strings are arrays” mean?
- A string (text) is just a row of letter boxes.
- Each letter sits in its own box, and Python numbers them.



```python
a = "HELLO WORLD 1234"
```

```python
print(a[1], end="")
print(a[5], end="")
print(a[12], end="")
```

### Slicing Strings
- You can return a range of characters by using the slice syntax.
- Specify the start index and the end index (not included), separated by a colon, to return a part of the string.
- Use **negative indexes** to start the slice from the **end of the string**

**Syntax** `string_name[<start_idx> : <end_idx> : <step>]` <br>
***Note:*** By default `<step>` value is `1`

```python
# Get the characters from position 2 to position 5 (not included)
# a = "HELLO WORLD 1234"
print(a[2:7])
```

```python
# Get the characters from the start to position 5 (not included):
print(a[:5]) # print(a[0:5])
```

```python
# Get the characters from position 2, and all the way to the end:
print(a[2:])
```

```python
print(a[:])
```

```python
print(a[::2])
```

```python
print(a[::-2])
```

```python
print(a[::-1])
```

```python
print(a[-5 :-2])
```

```python
print(a[-10:-2:2])
```

### String Format
Earlier we have learnt that one cannot combine strings and numbers like this:

```python
age = 36
#This will produce an error:
txt = "My name is John, I am " + age
print(txt)
```

But we can combine strings and numbers by using **`f-strings`** or the **`format()`** method!

#### F-Strings
To specify a string as an f-string, simply put an **f** in front of the string literal, and add curly brackets **{}** as placeholders for variables and other operations.

```python
age = 36
txt = f"My name is John, I am {age}"
print(txt)
```

***Note***: 
- A placeholder can contain ***variables, operations, functions, and modifiers*** to format the value.
- A placeholder can include a **modifier** to format the value.
- Perform a **math operation** in the placeholder, and return the result.

A modifier is included by adding a colon `:` followed by a legal formatting type, like `.2f` which means fixed point number with 2 decimals

```python
price = 59
txt = f"The price is {price} dollars"
print(txt)
```

```python
# Display the price with 2 decimals:

price = 59
txt = f"The price is {price:.2f} dollars"
print(txt)
```

```python
txt = f"The price is {20 * 60} dollars"
print(txt)
```

### Escape Character
- To insert characters that are illegal in a string, use an escape character.
- An escape character is a backslash `\` followed by the character you want to insert.

| Code | Result |
|---|---|
| `\'` | Single Quote |
| `\"` | Double Quotes |
| `\\` | Backslash |
| `\n`| New Line |
| `\r` | Carriage Return |
| `\t` | Tab |

```python
txt = "We are the so-called "Vikings" from the north." # error
```

```python
# The escape character allows you to use double quotes when you normally would not be allowed:

txt = "We are the so-called \"Vikings\" from the north."
print(txt)
```
