In python,string is a data type used to represent text.

strings are written in quotes because that is how python seperates that specific data type from other data types.

string indexing is a way to access certain strings from their positions. Positive indexing is when each digit of a string gets its own number, starting from 0, meanwhile negative indexing is the opposite, counting from the last digit, and going -1, -2, -3 etc.

Slicing is a way to extract a specific portion of a sequence.

The len() keyword controls the length of a certain string.

We need escape characters to include characters in a string that would otherwise not be able to include.
