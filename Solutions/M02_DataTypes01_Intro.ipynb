**1. What is a `data type`? Why does Python need `data types`?
Write the answer in the markdown cell, do not use print statements. Use bold, italics and highlight, where ever it is necessary**

A **data type** is a classification that tells the computer what kind of value a variable holds and how it can be used. 

*Python requires data types for several key reasons:*

**Memory Management:**
Different data types require different amounts of memory storage. Specifying the type helps Python efficiently manage and allocate the appropriate amount of memory

**Defining Valid Operations:**
Data types dictate what operations are permissible. For example, you can perform mathematical addition on two integers, but attempting to "add" an integer to a string typically results in a type error unless explicitly handled

**Interpretation and Clarity:**
Types provide clarity for the interpreter (and the programmer) on how to interpret data. The value 5 is treated as a number in mathematical operations, while the value "5" (a string) is treated as text

**Error Prevention:**
Enforcing type rules at runtime helps catch logical errors early, preventing unexpected behavior or program crashes that could arise from applying incorrect operations to data
