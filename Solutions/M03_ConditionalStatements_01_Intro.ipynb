Q1. **Mission Control --- Launch Decision System**
You are the launch officer, and your program must make the final call.
- Ask for the fuel level as a percentage and convert it to a number.
- Ask whether the weather is clear (the user types `yes` or `no`); use a string method so capital or small letters both work.
- Ask how many crew members are ready and convert it to a whole number.
- Print `GO FOR LAUNCH` only when the fuel is 90 or more, AND the weather is clear, AND at least 3 crew are ready (use `and`).
- If the fuel is from 70 up to 89 (and the other checks are fine), print `Fuel low - top up before launch`.
- In every other case, print `LAUNCH ON HOLD`.
- Extra challenge: after your decision, use separate `if` statements to print a line for each check that failed (for example, `Weather not clear`).

````python
fuel = float(input("Enter fuel level (%): "))
weather = input("Is the weather clear? (yes/no): ").strip().lower()
crew_ready = int(input("How many crew members are ready? "))

if fuel >= 90 and weather == "yes" and crew_ready >= 3:
    print("GO FOR LAUNCH")

elif fuel >= 70 and fuel <= 89 and weather == "yes" and crew_ready >= 3:
    print("Fuel low - top up before launch")

else:
    print("LAUNCH ON HOLD")
````

Q2. **Exoplanet Habitability Checker**
Astronomers want to know whether a newly found planet could support life.
- Ask for the planet's average surface temperature in degrees Celsius and convert it to a decimal number.
- Ask whether liquid water is present (`yes` or `no`).
- Print `Possibly habitable!` when the temperature is from 0 up to 50 AND water is present.
- Print `Right temperature, but no water` when the temperature is from 0 up to 50 but there is no water.
- Print `Too cold` when the temperature is below 0.
- Print `Too hot` for anything else.

````python
temp = float(input("Enter average surface temperature (°C): "))
water = input("Is liquid water present? (yes/no): ").strip().lower()

if 0 <= temp <= 50 and water == "yes":
    print("Possibly habitable!")
elif 0 <= temp <= 50 and water == "no":
    print("Right temperature, but no water")
elif temp < 0:
    print("Too cold")
else:
    print("Too hot")
````

Q3. **Mars Rover Battery Manager**
A Mars rover decides what to do based on how much battery it has left.
- Ask for the battery level as a percentage and convert it to a whole number.
- Print `Battery full - explore freely` when it is 80 or more.
- Print `Battery good - continue mission` when it is from 50 up to 79.
- Print `Battery low - return to base` when it is from 20 up to 49.
- Print `Critical - enter sleep mode` for anything lower.
- Then ask whether it is currently daytime (`yes` or `no`). Using a separate `if` statement, print `Solar panels charging` only when it is daytime.

````python
battery = int(input("Enter battery level (%): "))

if battery >= 80:
    print("Battery full - explore freely")
elif battery >= 50:
    print("Battery good - continue mission")
elif battery >= 20:
    print("Battery low - return to base")
else:
    print("Critical - enter sleep mode")

daytime = input("Is it daytime? (yes/no): ").strip().lower()

if daytime == "yes":
    print("Solar panels charging")
````
