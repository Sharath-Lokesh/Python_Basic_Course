Q16. Explain all the sets and frozen methods with an example for each.

copy returns a shallow copy, difference returns elements in a main set but not others, intersection returns all elements of existing sets, isdisjoint returns true if 2 sets have different elements, issubset returns true if the selcted set is in another set, issuperset returns true if a selected set contains another set, symetric_difference returns elements from a set out of 2 and not the other, union returns only the unique elements from 2 sets, add adds a singular element to a selected set, remove removes specified elements from a selected set, difference_update removes all the elements included in another set, discard removes an element, intersection_update only keeps the common elements found in 2 specified sets, pop removes an element and returns that element, symetric_difference_update only keeps the elements that are in 1 randomized set out of 2, and update adds a value or characteristic of choice to the set.

````python
car = {"Ford", "Mustang", 1964}
keys={"Ford","Mustang",1963}
car_copy = car.copy()
print (car_copy)
new_car = car.difference()
print (new_car)
model = car.intersection()
print (model)
color = car.isdisjoint("Ford")
print (color)
removed_year = car.issubset("Ford")
print (removed_year)
last_item = car.issuperset("Ford")
print (last_item)
car_keys=car.symmetric_difference("Ford")
print (car_keys)
car_values = car.union()
print (car_values)
car_items = car.add(1)
print (car_items)
car_1=car.remove(1964)
print (car)
car_2=car.difference_update()
print (car)
car_3=car.discard("Ford")
print (car_3)
car_4=car.intersection_update()
print (car_4)
car_5=car.pop()
car_6=car.symmetric_difference_update("Ford")
print (car_6)
car_7=car.update("car")
````

Q17. Explain all the dictionary methods with an example for each.

 clear completely wipes all values from a dictionary, copy returns a unconnected copy of a dictionary, fromkeys creates a new dictionary using an iterable, get returns the value of a key, items gives you all the values of a dictionary in the form of a tuple, key gives you only the keys of a dictionary, pop removes a certain key-value pair and give you that, popitem removes an item, and give you it in the form of a tuple, setdefault returns the value of a key if it is already in the dictionary. If the key does not exist, it inserts the key with a specified value and then returns that value, update adds more key-value pairs to the dictionary, and values gives you all the values in the dictionary

````python
car = {"brand": "Ford", "model": "Mustang", "year": 1964}
car.clear()
print(car)

car = {"brand": "Ford", "model": "Mustang"}
new_car = car.copy()
new_car["color"] = "red"
print(car)
print(new_car)

keys = ('key1', 'key2', 'key3')
value = 0
new_dict = dict.fromkeys(keys, value)
print(new_dict)

car = {"brand": "Ford", "model": "Mustang"}
print(car.get("model"))
print(car.get("price", "Not Found"))

car = {"brand": "Ford", "model": "Mustang"}
print(car.items())

car = {"brand": "Ford", "model": "Mustang"}
print(car.keys())

car = {"brand": "Ford", "model": "Mustang", "year": 1964}
removed_value = car.pop("model")
print(removed_value)
print(car)

car = {"brand": "Ford", "model": "Mustang", "year": 1964}
last_item = car.popitem()
print(last_item)
print(car)

car = {"brand": "Ford", "model": "Mustang"}
color = car.setdefault("color", "white")
brand = car.setdefault("brand", "Chevy")
print(color)
print(brand)
print(car)

car = {"brand": "Ford", "model": "Mustang"}
car.update({"color": "White", "year": 1964})
print(car)

car = {"brand": "Ford", "model": "Mustang"}
print(car.values())
````
