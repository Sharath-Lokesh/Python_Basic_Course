The difference between local and remote repositories, is that local repositories are private and only available to a few members of the owners choosing, meanwhile remote repositories are accessable by anyone.

git clone is a command used to completely copy the contents of a certain repository

e.g:   git clone<(repository url)>

The git branch <name> command creates a new development line, while git checkout <name> moves you into it. You use it to work on a specific feature in isolation. This is needed so you can experiment or fix bugs without breaking the main, stable code.

The git clone <url> command copies an existing project from a server to your computer. You use it by pasting the repository URL after the command in your ter
minal. It is needed to get a local, workable copy of a project's entire history and files.

The git add <file> command prepares specific changes to be saved. You use it by naming the files you've modified or using git add . for all files. This is needed as a "middle step" to pick exactly which changes belong in your next snapshot.

The git commit -m "message" command officially saves your staged changes. You use it by providing a short description of what you did. This is needed to create a permanent restore point in your project’s history.

The git log command shows a list of all past saves. You use it by typing the command to scroll through the project's timeline. It is needed to track who made changes, what they did, and when they did it.

The git status command shows which files are changed, staged, or untracked. You use it frequently to check your current progress. It is needed to ensure you don’t accidentally commit the wrong files or lose track of your work.

pull/merge request is a request made on a website like GitHub to merge your branch into the main project. You use it by clicking "New Pull Request" after pushing your code online. It is needed so teammates can review and approve your code before it becomes official.
