#In a match statement, what does the underscore case case _: do, and why should it always be placed last?

_ is the default case. It matches anything that has not matched an earlier case.
It should always be placed last because it matches all remaining values. If it came earlier, later cases would never be reached.

Q5. Start with this dictionary of surface gravity (in m/s²):
```python
gravity = {"Earth": 9.8, "Mars": 3.7, "Moon": 1.6}
```
Ask the user for a place name. Use a **nested if**: the *outer* if checks whether the name is a key in the dictionary (use the `in` keyword), and the *inner* if checks whether that place's gravity is less than 5. Print a suitable message for a low-gravity place, a high-gravity place, and for a place that is not in the dictionary.

````python
gravity = {"Earth": 9.8, "Mars": 3.7, "Moon": 1.6}

place = input("Enter a place name: ")

if place in gravity:
    if gravity[place] < 5:
        print(f"{place} is a low-gravity place.")
    else:
        print(f"{place} is a high-gravity place.")
else:
    print("That place is not in the dictionary.")
````

Q9. Ask the user to type a planet name in lower case. Use a **match** statement, grouping values with `|`, to print `Rocky planet` for mercury/venus/earth/mars, `Gas giant` for jupiter/saturn, `Ice giant` for uranus/neptune, and `Unknown planet` for anything else.

Q10. Start with these two values:
```python
fuel = 80
day = 4
```

````python
fuel = 80
day = 4

match day:
    case 1 | 2 | 3 | 4 | 5 if fuel >= 100:
        print("Weekday launch: GO")

    case 1 | 2 | 3 | 4 | 5 if fuel < 100:
        print("Weekday launch: HOLD - low fuel")

    case 6 | 7:
        pass  # No action needed because no launches occur on weekends
````
