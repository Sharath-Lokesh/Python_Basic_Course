# Python Project Assignment: GUI Calculator (tkinter)

## How to use this assignment
- Keep an untouched backup copy of `calculator.py` before you start.
- Answer the theory and reasoning questions in your own words. Add a new markdown cell under a question to write your answer.
- For the hands-on tasks: make the change in `calculator.py`, run it, then paste your changed code - **not the whole updated code** - (the function or the lines you changed) into the answer cell and add one line describing what now works differently. A screenshot of the result is welcome.
- You only need the Python tools already used in the calculator. No new loops or classes are required beyond what the app already contains.

## Difficulty levels
- Questions marked **[Core]** are the essential set.
- Questions marked **[Stretch]** go deeper.
- Questions marked **[Research]** expect you to look something up before answering.
